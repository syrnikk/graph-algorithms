```python

import math
import matplotlib.pyplot as plt
import graph


def read_data(filename):
    with open(filename, 'r') as f:
        data = f.read().splitlines()
    return [tuple(map(int, line.split())) for line in data]


def make_plot(vertices, title):
    # extract x and y coordinates into separate lists
    x = [vertex[0] for vertex in vertices]
    y = [vertex[1] for vertex in vertices]

    # create a new figure and axis
    fig, ax = plt.subplots()

    # plot the vertices as scatter points
    ax.scatter(x, y)

    # plot the connections between the vertices
    for i in range(len(vertices)):
        ax.plot([x[i], x[(i+1)%len(vertices)]], [y[i], y[(i+1)%len(vertices)]])

    # set title
    plt.title(title)    
    
    # show the plot
    plt.show()
    

def distance(a, b):
    dx = a[0] - b[0]
    dy = a[1] - b[1]
    return math.sqrt(dx*dx + dy*dy)


def total_distance(vertices):
    return sum(distance(vertices[i], vertices[(i+1) % len(vertices)]) for i in range(len(vertices)))
```

```python
vertices = read_data('lab6/examples/data.dat')
make_plot(vertices, 'Before simulated annealing algorithm')
print(f'Total distance: {total_distance(vertices)}')
```

```python
vertices = graph.simulated_annealing(vertices)
make_plot(vertices, 'After simulated annealing algorithm')
print(f'Total distance: {total_distance(vertices)}')
```
