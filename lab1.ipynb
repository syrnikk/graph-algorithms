```python
import numpy as np
import lab1.graph_reader_writer as graph_reader_writer
import lab1.graph_converter as graph_converter
import lab1.graph_generator as graph_generator
import lab1.graph_visualizer as graph_visualizer
```

```python
# reading from file
graph = None
with open("./lab1/examples/list.txt", "r") as f:
    graph = graph_reader_writer.read_list(f, v_sep=";")
    graph_reader_writer.print_list(graph)
matrix = graph_converter.convert_list_to_matrix(graph)
```

```python
# printing and converting in different formats
print("\nNeighbor matrix for graph above")

graph_reader_writer.print_nei_matrix(
    [i for i in range(np.shape(matrix)[0])], matrix)

matrix2 = graph_converter.convert_list_to_incidence(graph)
matrix2_shape = np.shape(matrix2)

print("\nIncidence matrix")
graph_reader_writer.print_incidence_matrix([i for i in range(matrix2_shape[0])], [
    i for i in range(matrix2_shape[1])], matrix2)
```

```python
display(graph_visualizer.GraphVisualizer(graph).get_image())
```

```python
with open("./lab1/examples/nei_matrix_2.txt") as f:
    mt = graph_reader_writer.read_nei_matrix(f, v=True, v_sep="t", n_sep=" ")
    graph_reader_writer.print_nei_matrix(mt[0], mt[1])
    print()
    li = graph_converter.convert_matrix_to_list(mt[1])
    graph_reader_writer.print_list(li)
    print()
    inc = graph_converter.convert_list_to_incidence(li)
    graph_reader_writer.print_incidence_matrix([], [], inc)
    print()
```
